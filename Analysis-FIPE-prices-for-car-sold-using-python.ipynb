```python
#boxplot for cheking outliers
plt.figure(figsize=(12,8),facecolor="pink")
pltn=1
for i in numcol:
    if pltn<=5:
        ax=plt.subplot(3,2,pltn)
        sns.boxplot(df[i],color="blue")
        plt.xlabel(i,fontsize=13)
    pltn=pltn+1
plt.tight_layout() 
```

```python
##distplot is also cheking outliers
plt.figure(figsize=(12,8),facecolor="pink")
pltn=1
for i in numcol:
    if pltn<=5:
        ax=plt.subplot(3,2,pltn)
        sns.distplot(df[i],color="blue")
        plt.xlabel(i,fontsize=13)
    pltn=pltn+1
plt.tight_layout() 
```

```python
#removing outliers
from scipy.stats import zscore
features=df[['engine_size','avg_price_brl']]
z=np.abs(zscore(features))
```

```python
z
```

```python
new_df=df[(z<3).all(axis=1)]
new_df
```

```python
# Checking the shape of dataset before and after removing outliers
print("Shape of dataset before removing outliers:", df.shape)
print("Shape of dataset after removing outliers:", new_df.shape)
```
